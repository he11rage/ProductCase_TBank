***Данные охватывают сезон аренды самокатов 2024 года (с апреля по октябрь 2024).***

```python
# График count_plot
def count_plot(data, x, title=None, xlabel=None, count_order=False, rotate=False):
    plt.figure(figsize=(10, 8))
    sns.set(style="whitegrid")
    plt.rcParams['axes.labelsize'] = 14
    plt.rcParams['xtick.labelsize'] = 12
    plt.rcParams['ytick.labelsize'] = 12
    if count_order:
        order = data[x].value_counts().index
        ax = sns.countplot(data, x=x, order=order, hue=x, legend=False, palette='blend:#7AB,#EDA')
    else:
        ax = sns.countplot(data, x=x, hue=x, legend=False, palette='blend:#7AB,#EDA')
    if rotate:
        ax.tick_params(axis='x', labelrotation=45)

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel('Количество')
```

```python
# Процент от общего числа
def count_percent(data, col, is_series=False):
    if is_series:
        filled = data.fillna('Unknown')
    else:
        if col is None:
            raise ValueError("Укажите имя столбца col для DataFrame.")
        filled = data[col].fillna('Unknown')
    counts = filled.value_counts()
    perc = (counts / counts.sum() * 100).round(1)
    return pd.DataFrame({'Количество': counts, 'Процент, %': perc})
```

## Анализ признаков

### Числовые признаки

```python
# Заполним пропуски в выплаченном кэшбеке/кэшбек бонусами нулями, так как отсутствие кэшбека означает что он равен нулю.
data[['loyalty_accrual_rub_amt', 'loyalty_accrual_bns_amt']] = data[['loyalty_accrual_rub_amt', 'loyalty_accrual_bns_amt']].fillna(0)
```

```python
# Отберем информативные числовые признаки
num_chars = data[['age', 'minute_cost', 'activation_cost', 'hold_amount', 'distance_km', 'book_time_zone_cd', 'nominal_price_rub_amt', 'loyalty_accrual_rub_amt', 'loyalty_accrual_bns_amt']]

# Посмотрим их характеристики
num_chars.describe().T
```

```python
# Посмотрим сколько отрицательных значений кешбека
less_zero_cashback = data[['loyalty_accrual_rub_amt']].query("loyalty_accrual_rub_amt < 0")
print(f"Кол-во отрицательных записей: {less_zero_cashback.value_counts().sum()}")
less_zero_cashback.hist(bins=10)
```

```python
# Заменим отрицательные значения нулями
data['loyalty_accrual_rub_amt'] = data['loyalty_accrual_rub_amt'].clip(lower=0)
```

```python
# Посчитаем кол-во аномальных значения для километража поездки
upper_limit = data['distance_km'].quantile(0.995)
print(upper_limit)
big_distance = data[['distance_km']].query(f"distance_km > {upper_limit}")
big_distance.value_counts().sum()
```

```python
# Удалим значения и выбросы выше 99.5% перцентиля
data = data.query(f"distance_km <= {upper_limit}")
data['distance_km'].hist(bins=10)
```

**Вывод:**

В целом числовые признаки с виду без видимых аномалий/выбросов, за исключением признаков:
- **Километраж поездки**: обнаружены экстремальные значения, включая максимальное 56 000 км. Значения, превышающие 99.5-й перцентиль, были удалены, чтобы сохранить естественное распределение данных.
- **Размер выплаченного кешбека**: обнаружены 7 отрицательных записей. Эти значения заменены на нули, так как отрицательный кешбек невозможен.

Пропуски в признаках "Размер выплаченного кешбека" и "Размер выплаченного кешбека бонусами" заполнены.

```python
# Построим матрицу корреляции для числовых признаков
plt.figure(figsize=(10, 8))
corr = num_chars.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title('Матрица корреляции')
plt.show()
```

**Вывод:** 
Исходя из матрицы корреляции можно сделать следующие выводы:
- Стоимость минуты, стоимость активации, зависят от часового пояса (региона), в котором арендуется самокат, чем больше часовой пояс/восточнее регион, тем ниже платежеспособность -> ниже цена
- Также вполне логично, что время поездки очень сильно коррелирует с километражом
- Размер кешбека зависит от суммы поездки
- Остальные признаки не показали значимых результатов корреляции

```python
# Переведем признаки из типо object в datetime
data['created_dttm'] = pd.to_datetime(data['created_dttm'])
data['book_start_dttm'] = pd.to_datetime(data['book_start_dttm'])
data['book_end_dttm'] = pd.to_datetime(data['book_end_dttm'])
data['local_book_start_dttm'] = pd.to_datetime(data['local_book_start_dttm'])
```

```python
# Выделим в отдельный датасет все дата объекты
date_data = data[['created_dttm', 'book_start_dttm', 'book_end_dttm']]
```

```python
date_data.info()
```

```python
start, end = data['created_dttm'].min(), data['created_dttm'].max()
period = end - start
f"Данные в датасете собраны за период с {start:%d.%m.%Y} по {end:%d.%m.%Y} ({period.days} дней)"
```

#### Активность клиентов по дням и часам

```python
days_map = {
    'Monday': 'Понедельник',
    'Tuesday': 'Вторник',
    'Wednesday': 'Среда',
    'Thursday': 'Четверг',
    'Friday': 'Пятница',
    'Saturday': 'Суббота',
    'Sunday': 'Воскресенье'
}

data['hour'] = data['local_book_start_dttm'].dt.hour
data['weekday'] = data['local_book_start_dttm'].dt.day_name().map(days_map)
data['month'] = data['local_book_start_dttm'].dt.month
```

```python
activity = data.groupby(['weekday', 'hour']).size().reset_index(name='count')
```

```python
order_days = ['Понедельник', 'Вторник', 'Среда', 'Четверг', 'Пятница', 'Суббота', 'Воскресенье']
activity['weekday'] = pd.Categorical(activity['weekday'], categories=order_days, ordered=True)
```

```python
pivot = activity.pivot(index='weekday', columns='hour', values='count')

plt.figure(figsize=(18,6))
sns.heatmap(pivot, cmap='crest')
plt.title('Активность пользователей по дням недели и часам', fontsize=14)
plt.xlabel('Часы суток')
plt.ylabel('День недели')
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
plt.show()
```

**Вывод**:
- Наивысшее пики активностей пользователей в будни в районе 5:00 часов утра и 14:00-17:00.
- В выходные активность меньше чем в будние дни и имеет более плавный график.

#### Активность по месяцам

```python
month_activity = data.groupby(['month']).size().reset_index(name='count')
month_map = {
    1 : 'Январь',
    2 : 'Февраль',
    3 : 'Март',
    4 : 'Апрель',
    5 : 'Май',
    6 : 'Июнь',
    7 : 'Июль',
    8 : 'Август',
    9 : 'Сентябрь',
    10 : 'Октябрь',
    11 : 'Ноябрь',
    12 : 'Декабрь'
}
month_activity['month_name'] = month_activity.month.map(month_map)
month_activity
```

```python
plt.figure(figsize=(10, 8))
sns.set_style("whitegrid")
sns.barplot(data=month_activity, x='month_name', y='count', hue='month_name', legend=False, palette='blend:#7AB,#EDA')
plt.xlabel('Месяц')
plt.ylabel('Количество поездок')
plt.title('Количество поездок по месяцам')
plt.rcParams['axes.labelsize'] = 14
plt.rcParams['xtick.labelsize'] = 12
plt.rcParams['ytick.labelsize'] = 12
```

**Вывод**:
- Наибольший пик активности клиентов попадает на август и сентябрь.
- В начале сезона наблюдается наименьшая активность.

# Доля рынка

Перечислим популярные сервисы и их информацию о количестве поездок за сезон.
Количество поездок за сезон 2024 по официальным данным:
- Whoosh - 149.7 млн.
- МТС Юрент - 100.3 млн.
- Яндекс GO - 86.4 млн.
- Сервис Т-Банк - 396 тыс.

**Построим круговой график с долями рынка:**

```python
market_share = {
    'Service' : ['Whoosh', 'МТС Юрент', 'Яндекс GO', 'Т-Банк'], 
    'trips_count': [149700000, 100300000, 86400000, 396000]
}
df = pd.DataFrame(market_share)
df
```

```python
plt.figure(figsize=(8, 8))
colors = ['#82aeb9', '#aac0b3', '#d3d1ad', '#ebdcaa']
plt.pie(
    df['trips_count'], 
    labels=df['Service'], 
    autopct='%1.2f%%', 
    startangle=140,
    colors=colors
)
plt.title('Доля рынка по количеству поездок за сезон 2024')
plt.show()
```
